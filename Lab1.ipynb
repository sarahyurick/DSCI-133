```python
def square(x):
    return x*x
square(5)
```

* You can also pass function to function

```python
def sum_of_something(x, y, f):
    return f(x)+f(y)
sum_of_something(1,2,square)
```

Apart from the normal arguments, we can have some other types of arguments  
**default arguments, positional arguments and keyword arguments**.   
Default arguments are given a default value, it doesn't have to be given a value when calling the function.  
Positional arguments are stored in a tuple, and keyword arguments in a dictionary.   
Note the "starred" syntax.  
Notice that, you can only define the arguments in the order of **normal arguments, default arguments, positional arguments and keyword arguments**

```python
def f(a,b,default=0,*posargs,**kwargs):
    print "got",a,b,default,posargs, kwargs
    return a
print f(1,2)
print f(1,2,3)
print f(1,2,3,4,5)
print f(1,2,3,4,5, d=1, c=2)
```

### Sum up
* Let's mine some information from Hamlet.txt

```python
print len(hamlettokens)
```

* Convert all words into lower-cases, and count the appearance of a specific word

```python
hamletlctokens=[word.lower() for word in hamlettokens]
hamletlctokens.count("to")
```

Let's find the unique token set, by using the ``set`` function

```python
uniquetokens=set(hamletlctokens)
```

```python
len(uniquetokens)
```

```python
tokendict={}
for ut in uniquetokens:
    tokendict[ut]=hamletlctokens.count(ut)
```

Let's find the 20 most used words

```python
sorted?
```

```python
top20=list(sorted(tokendict.iteritems(), key=lambda (k,v): v, reverse=True))[:20]
print top20
```

```python
pos = np.arange(len(top20))
plt.bar(pos, [e[1] for e in top20]);
plt.xticks(pos+0.4, [e[0] for e in top20]);
plt.ylabel("counts")
plt.xlabel("word")
```
