## Filtering

```python
df.query("rating>4") #it creats another DataFrame instance
```

```python
df.query('year < 0 & rating >4') # and 
```

```python
df.query('rating ==4.03 | rating==4.02') # or
```

```python
df[(df.year < 0) & (df.rating > 4)] #Another way of query
```

We notice there's an error, that's because we are trying to convert a "None" into "int"

```python
df[df.rating_count=='None']
```
